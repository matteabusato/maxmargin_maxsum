```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
base_path = "results/"
file_name = base_path + "results_forced_exponential.txt"

df = pd.read_csv(file_name, sep="\\s+", engine='python')
df.to_csv(base_path + "forced_exponential.csv", index=False)
```

```python
# SQUARED
base_path = "results/"
file_name = base_path + "forced_squared.csv"
df = pd.read_csv(file_name)

M_to_alpha = {
    250: 0.25,
    300: 0.30,
    350: 0.35,
    400: 0.40,
    500: 0.50,
    600: 0.60,
    700: 0.70,
}

average_deltas = []

for M, alpha in M_to_alpha.items():
    subset = df[(df["M"] == M) & (df["Delta_max"] >= -3)]
    avg = subset["Delta_max"].mean()
    average_deltas.append((alpha, avg))

average_deltas.sort()
alphas, delta_means = zip(*average_deltas)

plt.figure(figsize=(8, 5))
plt.plot(alphas, delta_means, marker='o', linestyle='-', color='blue', linewidth=2, markersize=7)

plt.xlabel("Patterns per synapse $\\alpha$", fontsize=12)

plt.ylabel("Stability margin $\\Delta^*$", fontsize=12)

plt.title("Stability margin $\\Delta^*$ vs. Patterns per synapse $\\alpha$ - Squared external field", fontsize=14)

plt.grid(True, linestyle=':', alpha=0.7)

# Set limits for x and y axis if desired, based on typical ranges seen in such plots
# (You might need to adjust these based on your actual data values to avoid cutting off data)
plt.xlim(0.2, 0.75)
plt.ylim(-2, 35)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Data Processing for SQUARED ---
base_path = "results/"
file_name_squared = base_path + "forced_squared.csv"
df_squared = pd.read_csv(file_name_squared)

M_to_alpha_squared = {
    250: 0.25,
    300: 0.30,
    350: 0.35,
    400: 0.40,
    500: 0.50,
    600: 0.60,
    700: 0.70,
}

average_deltas_squared = []

for M, alpha in M_to_alpha_squared.items():
    subset = df_squared[(df_squared["M"] == M) & (df_squared["Delta_max"] >= -3)]
    avg = subset["Delta_max"].mean()
    average_deltas_squared.append((alpha, avg))

average_deltas_squared.sort()
alphas_squared, delta_means_squared = zip(*average_deltas_squared)

# --- Data Processing for EXPONENTIAL ---
file_name_exponential = base_path + "forced_exponential.csv"
df_exponential = pd.read_csv(file_name_exponential)

M_to_alpha_exponential = {
    300: 0.30,
    400: 0.40,
    500: 0.50,
    600: 0.60,
    700: 0.70,
}

average_deltas_exponential = []

for M, alpha in M_to_alpha_exponential.items():
    subset = df_exponential[(df_exponential["M"] == M) & (df_exponential["Delta_max"] >= -3)]
    avg = subset["Delta_max"].mean()
    average_deltas_exponential.append((alpha, avg))

average_deltas_exponential.sort()
alphas_exponential, delta_means_exponential = zip(*average_deltas_exponential)

# --- Plotting both on the same graph ---
plt.figure(figsize=(10, 6)) # Slightly larger figure for better readability with two lines

# Plot for SQUARED
plt.plot(alphas_squared, delta_means_squared,
         marker='o', linestyle='-', color='blue', linewidth=2, markersize=7, label='Squared external field')

# Plot for EXPONENTIAL
plt.plot(alphas_exponential, delta_means_exponential,
         marker='o', linestyle='-', color='red', linewidth=2, markersize=7, label='Exponential external field') # Changed marker, linestyle, and color for differentiation

plt.xlabel("Patterns per synapse $\\alpha$", fontsize=12)
plt.ylabel("Stability margin $\\Delta^*$", fontsize=12)
plt.title("Stability margin $\\Delta^*$ vs. Patterns per synapse $\\alpha$", fontsize=14)

plt.grid(True, linestyle=':', alpha=0.7)

# Set limits for x and y axis
plt.xlim(0.2, 0.75)
plt.ylim(-2, 35)

plt.legend(fontsize=10) # Add a legend to distinguish the plots
plt.tight_layout()
plt.show()
```
